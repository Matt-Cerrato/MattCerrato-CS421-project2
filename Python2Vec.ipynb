```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

from mine import aggregate
from remove_comments import do_file

aggregate("./Repos","totalCode.txt")
# do_file("totalCode.txt")
```

```python
import re
num_lines = 0
word_list = []
tokenized_list =[]
words = ""
with open("totalCode.txt,strip",'r') as code_list:
    for line in code_list:
        if len(line) > 0:
            words += line
            num_lines+=1
words.lower()
word_list = re.split(r'\s+|\,|\;|\:|\(|\)|\#|\'|\"',words)
for word in word_list:
    if len(word) > 0:
        tokenized_list.append(simple_preprocess(word))
print("number of lines of code: " + str(num_lines))
```

```python
print(tokenized_list[0:30])
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(tokenized_list)
```

```python
model.train(tokenized_list, total_examples=model.corpus_count, epochs=model.epochs)
```
